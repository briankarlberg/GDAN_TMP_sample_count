```python
metrics = pd.read_csv('../Figure_3/classifier_metrics_20210710/big_results_matrix.tsv',sep='\t')
```

```python
metrics_SK = metrics[metrics.model.str.contains('skgrid')]
```

```python
# Model extraction for skgrid
import pandas as pd
import argparse
# metrics = pd.read_csv('../Figure_3/classifier_metrics_20210710/big_results_matrix.tsv',sep='\t')
# metrics_SK = metrics[metrics.model.str.contains('skgrid')]
c_list=metrics_SK.cohort.unique()
store_dict = {}
store_dict_26_by_3 = {'Cohort':[], 'Model':[], 'Features':[]}
for c in c_list: # for each cancer
    print(c)
    CancerFrame = metrics_SK[metrics_SK.cohort == c] # make cancer frame 
    cohort_F1 = CancerFrame[CancerFrame.performance_metric.str.contains('overall_weighted_f1')].copy() # filter overall F1
    cohort_F1.reset_index(inplace=True,drop = True)
    cohort_F1.sort_values('Mean',ascending= False,inplace = True) # Best sk_grid models for this cancer
    cohort_F1.reset_index(inplace=True,drop = True)
    subset = cohort_F1[['model','Mean']].copy() # Just the model and mean score columns
    max_id = cohort_F1.Mean.idxmax()
    best_cohort_model = cohort_F1.loc[max_id,'model']
    features = cohort_F1.loc[max_id,'featureID']
    bcm = subset.loc[max_id,:] # Best cohort model
    
    max_plus_one = max_id + 1
    second_best_cohort_model = cohort_F1.loc[max_plus_one,:]
    sbcm = subset.loc[max_plus_one,:]
    
#     print(bcm)
#     print(sbcm)
    store_dict[c] = (best_cohort_model)
#     if key in case_list:
#         case_list[key1].append(value)
#     else:
#         case_list[key1] = [value]
    store_dict_26_by_3['Cohort'].append(c)
    store_dict_26_by_3['Model'].append(best_cohort_model)
    store_dict_26_by_3['Features'].append(features)
```

```python
three_cols = pd.DataFrame(data = store_dict_26_by_3, index = list(range(0,26)))
```

```python
three_cols.to_csv('sk_grid_best_models_20210710.tsv', sep = '\t', index = False)
```
